**Clustering is an unsupervised machine learning technique used to group similar data points into clusters. Unlike classification, clustering does not require labeled data. It identifies inherent patterns or groupings within the data based on certain similarity metrics.**

```python
from sklearn.datasets import make_blobs, make_moons
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x,y = make_blobs(n_features=2, n_samples=5000,centers=4)
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(x)
```

```python
predict = kmeans.predict(x)
```

```python
plt.scatter(x[:,0], x[:,1],c=predict)
```

```python
x, y = make_moons(n_samples=1000, noise=0.06)
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(x)
```

```python
predict = kmeans.predict(x)
```

```python
plt.scatter(x[:,0], x[:,1],c=predict)
```

Agglomerative Clustering is a type of hierarchical clustering that follows a bottom-up approach. It begins with each data point as an individual cluster and then successively merges the closest clusters until a single cluster (or a predefined number of clusters) is formed.

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ag1 = AgglomerativeClustering(linkage='single')
```

```python
ag1.fit(x)
```

```python
ag1.labels_
```

```python
plt.scatter(x[:,0], x[:,1],c=ag1.labels_)
```
