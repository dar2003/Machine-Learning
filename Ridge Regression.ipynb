**Ridge Regression is a type of linear regression that includes a regularization term to prevent overfitting and improve the model's generalization. It is particularly useful when there is multicollinearity (high correlation between independent variables) in the data.**


```python
from sklearn.datasets import make_regression
x,y = make_regression(n_features=1, noise=5, n_samples=10000)
import matplotlib.pyplot as plt
plt.scatter(x,y,s=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import Ridge
reg = Ridge()
reg.fit(X_train,y_train)
```

```python
reg.predict(X_test)
```

```python
reg.score(X_test, y_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(x)
```

```python
plt.scatter(x,y,label='Actual',s=2)
plt.scatter(x,y_pred,label='Predicted',s=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha=100)
reg.fit(X_train,y_train)
reg.score(X_test, y_test)
```

```python
y_pred = reg.predict(x)
```

```python
plt.scatter(x,y,label='Actual',s=2)
plt.scatter(x,y_pred,label='Predicted',s=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
reg.coef_
```

```python
reg = Ridge(alpha=100000)
reg.fit(X_train,y_train)
reg.score(X_test, y_test)
```

```python
y_pred = reg.predict(x)
```

```python
plt.scatter(x,y,label='Actual',s=2)
plt.scatter(x,y_pred,label='Predicted',s=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
reg.coef_
```

```python
reg = Ridge(alpha=10000000)
reg.fit(X_train,y_train)
reg.score(X_test, y_test)
```

```python
y_pred = reg.predict(x)
plt.scatter(x,y,label='Actual',s=2)
plt.scatter(x,y_pred,label='Predicted',s=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
reg.coef_
```

**Advantage of increasing alpha**
Coefficient decreases and comes near to zero 
**Disadvantage of increasing alpha**
score also decreases
